```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

response = requests.get("https://en.wikipedia.org/wiki/The_Economist_Democracy_Index#List_by_country")
soup = BeautifulSoup(response.text, "html.parser")
content = soup.find("table", {"class" : "wikitable sortable sticky-header col2left"})
```

```python
columns = ['Rank', 'Δ Rank', 'Country', 'Regime type', 'Overall score', 'Δ Score','Electoral process and pluralism',
           'Functioning of government', 'Political participation', 'Political culture', 'Civil liberties']

values = []
for tr in content.find_all("tr"):
  temp = []
  for td in tr.find_all("td"):
    temp.append(td.text.removesuffix("\n"))
  values.append(temp)

print(values[3:])
print(columns)
data = pd.DataFrame(values[3:], columns=columns)
data.drop(['Δ Rank', 'Regime type', 'Δ Score'], axis=1)
data.to_csv("Economist_Democracy_Index_Components.csv")
```
